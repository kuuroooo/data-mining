```python
#CLEANING DATA
import pandas as pd

# 1. Define expected ranges for each variable.
#    Format: variable_name: (min_value, max_value)
expected_ranges = {
    'mood': (1, 10),
    'circumplex.valence': (-2, 2),
    'circumplex.arousal': (-2, 2),
    'activity': (0, 1),
    # If you have more variables, add them here:
    # 'screen': (0, 1440),  # e.g., if screen usage is in minutes
    # 'appCat.social': (0, 1440), 
    # 'call': (0, 1),       # if call is an indicator
    # 'sms': (0, 1),
    # etc.
}

# 2. For each variable in expected_ranges, find rows that are out of range
outliers_list = []  # to store all out-of-range rows

for var_name, (min_val, max_val) in expected_ranges.items():
    # Filter dataset rows that match this variable
    mask_var = (df['variable'] == var_name)
    
    # Among these rows, find those whose value is < min_val or > max_val
    mask_outlier = (df['value'] < min_val) | (df['value'] > max_val)
    
    # Combine the two masks to find out-of-range rows for this variable
    outliers = df[mask_var & mask_outlier]
    
    if not outliers.empty:
        print(f"Variable '{var_name}': Found {len(outliers)} out-of-range rows.")
        outliers_list.append(outliers)

# 3. Combine all out-of-range rows (if you want a single dataframe of outliers)
if outliers_list:
    df_outliers = pd.concat(outliers_list, ignore_index=True)
    print("\nSummary of out-of-range rows:")
    print(df_outliers.head())
else:
    print("No out-of-range values found based on the defined expected ranges.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# === CONFIGURABLE PARAMETERS ===
csv_file_path = 'dataset_mood_smartphone.csv'   # <-- Change this to your actual file path
target_variable = 'mood'   # <-- Change this to the variable you want to visualize

# === LOAD DATA ===
print("Loading data...")
df = pd.read_csv(csv_file_path, parse_dates=['time'])

# === FILTER BY VARIABLE ===
print(f"Filtering data for variable: {target_variable}")
df_filtered = df[df['variable'] == target_variable]

# === PLOT OVERLAYED GRAPH ===
print("Generating overlay plot...")

plt.figure(figsize=(12, 6))

for user_id, group in df_filtered.groupby('id'):
    group = group.sort_values(by='time')
    plt.plot(group['time'], group['value'], label=f'User {user_id}', alpha=0.8)

plt.title(f'Overlayed Plot for Variable: {target_variable}')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend(title='User ID', loc='best', fontsize='small')
plt.grid(True)
plt.tight_layout()

# === SAVE OR SHOW ===
plt.savefig('overlay_plot.png')  # Or use plt.show() if you want to display it directly
print("Overlay plot saved as 'overlay_plot.png'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# === CONFIGURABLE PARAMETERS ===
csv_file_path = 'dataset_mood_smartphone.csv'   # <-- Change this to your actual file path
target_variable = 'appCat.communication'   # <-- Change this to the variable you want to visualize

# === LOAD DATA ===
print("Loading data...")
df = pd.read_csv(csv_file_path, parse_dates=['time'])

# === FILTER BY VARIABLE ===
print(f"Filtering data for variable: {target_variable}")
df_filtered = df[df['variable'] == target_variable]

# === CALCULATE STATISTICS ===
average_value = df_filtered['value'].mean()
max_value = df_filtered['value'].max()
min_value = df_filtered['value'].min()
print(f"Stats for '{target_variable}': Avg={average_value:.2f}, Max={max_value:.2f}, Min={min_value:.2f}")

# === PLOT OVERLAYED GRAPH ===
print("Generating overlay plot...")
plt.figure(figsize=(12, 6))

# Plot each user’s time series
for user_id, group in df_filtered.groupby('id'):
    group = group.sort_values(by='time')
    plt.plot(group['time'], group['value'], label=f'User {user_id}', alpha=0.8)

# Plot average line
plt.axhline(y=average_value, color='black', linestyle='--', linewidth=2, label=f'Average: {average_value:.2f}')
plt.text(df_filtered['time'].min(), average_value, f'Avg: {average_value:.2f}', color='black', fontsize=10,
         ha='left', va='bottom', backgroundcolor='white')

# Plot max line
plt.axhline(y=max_value, color='red', linestyle=':', linewidth=2, label=f'Max: {max_value:.2f}')
plt.text(df_filtered['time'].min(), max_value, f'Max: {max_value:.2f}', color='red', fontsize=10,
         ha='left', va='bottom', backgroundcolor='white')

# Plot min line
plt.axhline(y=min_value, color='blue', linestyle=':', linewidth=2, label=f'Min: {min_value:.2f}')
plt.text(df_filtered['time'].min(), min_value, f'Min: {min_value:.2f}', color='blue', fontsize=10,
         ha='left', va='top', backgroundcolor='white')

# Final plot setup
plt.title(f'Overlayed Plot for Variable: {target_variable}')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend(title='User ID / Stats', loc='best', fontsize='small')
plt.grid(True)
plt.tight_layout()

# === SAVE OR SHOW ===
plt.savefig('overlay_plot_with_avg_min_max.png')  # You can use plt.show() instead
print("Overlay plot with average, max, and min saved as 'overlay_plot_with_avg_min_max.png'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Define which variables you want to visualize
#    (Adjust this list to match the variables actually in your dataset)
variables_of_interest = [
    'mood',
    'circumplex.valence',
    'circumplex.arousal',
    'activity',
    'screen',
    'call',
    'sms',
    'appCat.social',
    'appCat.entertainment'
]

# 2. Loop over each variable and plot a histogram
for var_name in variables_of_interest:
    # Filter to rows for this variable
    df_var = df[df['variable'] == var_name]

    # Skip if empty
    if df_var.empty:
        print(f"No rows found for variable '{var_name}'. Skipping...")
        continue

    # Convert to numeric if necessary
    values = pd.to_numeric(df_var['value'], errors='coerce').dropna()

    # Plot histogram
    plt.figure()
    values.hist(bins=30)  # 30 bins is arbitrary; adjust as needed
    plt.title(f"Histogram of '{var_name}'")
    plt.xlabel(var_name)
    plt.ylabel("Frequency")
    plt.show()

    # (Optional) Also plot a box plot to see outliers/percentiles
    plt.figure()
    plt.boxplot(values, vert=False)
    plt.title(f"Box Plot of '{var_name}'")
    plt.xlabel(var_name)
    plt.show()
```
